```python
# CARREGANDO AS BIBLIOTECAS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CRIANDO O OBJETO DF A PARTIR DA LEITURA DO DATAFRAME IMIGRANTES_CANADA
df = pd.read_csv('imigrantes_canada.csv')
```

```python
# ESTE CODIGO DEFINE COMO INDICE DO DF A COLUNA PAIS
# O ARGUMENTO INPLACE = TRUE MODIFICA O DF DIRETAMENTE SEM CRIAR UMA COPIA
df.set_index('País', inplace=True)
```

```python
# CONFIGURANDO O TEMA PADRÃO DOS GRÁFICOS GERADOS COM A BIBLIOTECA SNS
sns.set_theme()
```

```python
# ESTE CODIGO SELECIONA OS 10 PAISES COM OS MAIORES VALORES NA COLUNA TOTAL
top_10 = df.sort_values('Total', ascending=False).head(10)
```

```python
# VISUALIZANDO O DATAFRAME
top_10
```

```python
# CRIANDO GRÁFICOS DE BARRAS HORIZONTAL
sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')
```

```python
# AQUI FOI AJUSTADO OS ROTULOS DOS EIXOS NO GRÁFICO DE BARRAS HORIZONTAL MOSTRANDO OS PAISES COM MAIOR IMIGRAÇÃO 
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set(title="Países com maior imigração para o Canadá\n1980 a 2013",
       xlabel='Número de Imigrantes',
        ylabel='')

plt.show()
```

```python
# GRAFICO DE BARRAS HORIZONTAL COM CONFIGURAÇÕES ADICIONAIS
fig, ax = plt.subplots(figsize=(8, 4))

ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('', fontsize=14)

plt.show()
```

```python
# ESTE CODIGO CRIA UMA FUNÇÃO QUE GERA UM GRÁFICO DE BARRAS HORIZONTAL USANDO PALETA DE CORES
# FORNECEDIAS COMO ARGUMENTO
# PARA ENCONTRAR O NOME DA PALETA E NECESSÁRIO CONSULTAR A DOCUMENTAÇÃO DA BIBLIOTECA SEABORN
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
    ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
    ax.set_xlabel('Número de imigrantes', fontsize=14)
    ax.set_ylabel('')
    plt.show()
```

```python
# CHAMANDO A FUNÇÃO E PASSANDO A PALETA DE CORES 'ROCKET'
gerar_grafico_paleta('rocket')
```

```python
# CHAMANDO A FUNÇÃO E PASSANDO A PALETA DE CORES 'COOLWARM'
gerar_grafico_paleta('coolwarm')
```

```python
# CHAMANDO A FUNÇÃO E PASSANDO A PALETA DE CORES 'TAB10'
gerar_grafico_paleta('tab10')
```

```python
# DEFININDO O THEMA COMO 'DARK'
# CHAMANDO A FUNÇÃO E PASSANDO A PALETA DE CORES 'TAB10'
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python
# DEFININDO O THEMA COMO 'WHITEGRID'
# CHAMANDO A FUNÇÃO E PASSANDO A PALETA DE CORES 'TAB10'
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10')
```

```python
# DEFININDO O THEMA COMO 'WHITE'
# CHAMANDO A FUNÇÃO E PASSANDO A PALETA DE CORES 'TAB10'
sns.set_theme(style='white')
gerar_grafico_paleta('tab10')
```

```python
# DEFININDO O THEMA COMO 'TICKS' O QUE TAMBÉM ADICIONA TICKS NAS BARRAS HORIZONTAL
# CHAMANDO A FUNÇÃO E PASSANDO A PALETA DE CORES 'TAB10'
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10')
```

```python
# O CODIGO GERA O GRAFICO DE BARRAS HORIZONTAL ELENCANDO OS PAISES COM MAIOR IMIGRAÇÃO
# UTILIZANDO A PALETA DE CORES 'TAB10'
# AJUSTANDO DETALHES DO ESTILO, COMO TITULO, ROTULOS DOS EIXOS E REMOÇÃO DE BORDAS
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette='tab10')

ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')
sns.despine()

plt.show()
```
